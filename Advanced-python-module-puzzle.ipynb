```python
# In this example we have folder called 'unzip_me_for_instructions' which is a zip file .So we have to extract that zip file into a 
# new folder (Puzzle). When you extract that file there are four sub_folders in it and those sub folders contains text files. One of the 
# text file contains a number of pattern xxx-xxx-xxxx so now we have to find that pattern using os ,shutil and re (regex) methods




import shutil
folder_name='Puzzle'
file_to_unzip='C:\\Users\\DELL\\Desktop\\unzip_me_for_instructions.zip'
```

```python
shutil.unpack_archive(file_to_unzip,folder_name,'zip')
```

```python
import os
import re
file_path='C:\\Users\\DELL\\Puzzle'
```

```python
for folders,sub_folders,files in os.walk(file_path):
    print(f'Currently looking at : {folders}')
    print('\n')
    print('The sub folders are:')
    for sub_fold in sub_folders:
        print(f'\t The sub folders: {sub_fold}')
    print('\n')
    print('The files are: ')
    phone_numbers=[ ]
    for txt_file in files:
        print(f'\t The files: {txt_file}')
        f = open(folders+'\\'+txt_file,'r')
        text=f.read()
        for match in re.finditer(r'\d{3}-\d{3}-\d{4}',text):
            print(f'I have found a match in {txt_file} file : {match.group()} ')
    print('\n')
```
